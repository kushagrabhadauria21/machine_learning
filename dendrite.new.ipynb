```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np

iris = load_iris()
X = iris.data[:, 0].reshape(-1, 1)  # select sepal_length as feature and reshape to 2D array
y = iris.data[:, 3]  # select petal_width as target

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Define regression model and train on training data
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Make predictions on testing data and evaluate model
y_pred = regressor.predict(X_test)
y_pred
```

```python
from sklearn.datasets import load_iris
from sklearn.impute import SimpleImputer
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np

iris = load_iris()  # Call the function to load the dataset

# Create a list of the numerical feature names
num_feature_names = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']

# Impute missing values for the numerical features
imputer = SimpleImputer(missing_values=-1, strategy='mean')
for feature_name in num_feature_names:
    feature_values = iris.data[:, iris.feature_names.index(feature_name)].reshape(-1, 1)
    iris.data[:, iris.feature_names.index(feature_name)] = imputer.fit_transform(feature_values).ravel()

# Tokenize and hash the "species" text feature
vectorizer = CountVectorizer(token_pattern=r'\b\w+\b')
X_species = vectorizer.fit_transform(iris.target_names[iris.target])
hashed_feature_names = vectorizer.get_feature_names()

# Combine the numerical and text features into a single feature matrix
X_num = iris.data
X_all = np.hstack((X_num, X_species.toarray()))

# Print the final feature matrix
print("Final feature matrix:\n", X_all)
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()

# Create a Pandas DataFrame to store the dataset
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                  columns= iris['feature_names'] + ['target'])

# Compute the correlation coefficients between each feature and the target variable
correlations = df.corr()['target'][:-1]

# Select the top n features based on their correlation coefficients
n = 2
top_n_features = correlations.abs().nlargest(n).index.tolist()

# Extract the selected features from the dataset
X_reduced = df[top_n_features].values
X_reduced
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.ensemble import ExtraTreesRegressor

iris = load_iris()

# Create a Pandas DataFrame to store the dataset
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                  columns= iris['feature_names'] + ['target'])

# Split the dataset into input features and target variable
X = df.drop('target', axis=1)
y = df['target']

# Train an ExtraTreesRegressor model to compute feature importance
model = ExtraTreesRegressor(n_estimators=100, random_state=42)
model.fit(X, y)

# Select the top n features based on their feature importance scores
n = 2
top_n_features = np.argsort(model.feature_importances_)[::-1][:n]

# Extract the selected features from the dataset
X_reduced = X.iloc[:, top_n_features].values
X_reduced
```
